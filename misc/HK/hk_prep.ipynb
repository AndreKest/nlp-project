```python
import spacy
from spacy import displacy
from spacy.attrs import ORTH, NORM
import json
from bs4 import BeautifulSoup
```

```python
jsonObjects = []
with open("2019-02-19_oldp_cases.json") as jsonFile:
# with open("onecase.json") as jsonFile:
  for jsonObj in jsonFile:
    jsonObjects.append(json.loads(jsonObj))
  jsonFile.close()
```

```python
print(jsonObjects[6]["content"])
```

```python
nlp = spacy.load("de_core_news_lg")
```

```python
special_case = [{ORTH: "i.", NORM: "in"}, {ORTH: "H.", NORM: "Höhe"}, {ORTH: "v.", NORM: "von"}]
nlp.tokenizer.add_special_case('i.H.v.', special_case)

special_case = [{ORTH: "Ziff.", NORM: "Ziffer"}]
nlp.tokenizer.add_special_case('Ziff.', special_case)

special_case = [{ORTH: "i.", NORM: "im"}, {ORTH: "S.", NORM: "Sinne"}, {ORTH: "d.", NORM: "des"}]
nlp.tokenizer.add_special_case('i.S.d.', special_case)

special_case = [{ORTH: "i.", NORM: "in"}, {ORTH: "V.", NORM: "Verbindung"}, {ORTH: "m.", NORM: "mit"}]
nlp.tokenizer.add_special_case('i.V.m.', special_case)

special_case = [{ORTH: "Az.", NORM: "Aktenzeichen"}]
nlp.tokenizer.add_special_case('Az.', special_case)

special_case = [{ORTH: "Rn.", NORM: "Randnummer"}]
nlp.tokenizer.add_special_case('Rn.', special_case)

special_case = [{ORTH: "Aufl.", NORM: "Auflage"}]
nlp.tokenizer.add_special_case('Aufl.', special_case)

special_case = [{ORTH: "Abs.", NORM: "Absatz"}]
nlp.tokenizer.add_special_case('Abs.', special_case)
```

```python
doc_num = 12326
soup = BeautifulSoup(jsonObjects[doc_num]["content"], "lxml")
print(jsonObjects[doc_num]["id"])
# for i in soup.find_all('span'):
#   i.append(" ")
print(soup)
```

```python
doc = nlp(soup.text)


for sent in list(doc.sents):
  print(str(sent).strip("\n "))
  print("\n -------")
```

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)


displacy.render(doc, style="dep")
```

```python
text = "The formula is no. 45. This num. represents the chemical properties."

nlp = spacy.load("en_core_web_sm")
doc = nlp(text)
print("Before:", [sent.text for sent in doc.sents])


def set_custom_boundaries(doc):
    pattern_a = ['no', 'num']
    for token in doc[:-1]:
        if token.text in pattern_a and doc[token.i + 1].text == '.':
            doc[token.i + 2].is_sent_start = False
    return doc
    


nlp.add_pipe("test", before="parser")
doc = nlp(text)
print("After:", [sent.text for sent in doc.sents])
```
